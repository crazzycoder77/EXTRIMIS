```python
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer


```

```python
from selenium import webdriver
import requests
from selenium.webdriver.support.ui import WebDriverWait 
from time import sleep
from selenium.webdriver.common.by import By 
import datetime
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.common.keys import Keys 

waitTime = "2 sec"

browser = webdriver.Chrome(r'E:\chromedriver.exe')
browser.get('https://web.whatsapp.com/')
target = '"Jio2"'
x_arg = '//span[contains(@title,' + target + ')]'
  
def getReply(chatbot, metadata):
    #a = datetime.datetime.now()
    text_box = browser.find_element_by_class_name("_2S1VP")
    #text_box.send_keys("Wait my cuurent response time is " + str(waitTime) + "\n")
    metadata = metadata.split('\n')[0]
    #text_box.send_keys("***Debugging Mode is On... Input('"+ metadata +"')***\n")
    response = str(chatbot.get_response(metadata))
    text_box.send_keys(response + "\n")
    #b = datetime.datetime.now()
    #waitTime = b - a

    
bot_users = {} # A dictionary that stores all the users that sent activate bot 

wait = WebDriverWait(browser, 600)


while True:
    unread = browser.find_elements_by_class_name("OUeyt") # The green dot tells us that the message is new
    name,message  = '',''
    if len(unread) > 0:
        ele = unread[-1]
        action = webdriver.common.action_chains.ActionChains(browser)
        action.move_to_element_with_offset(ele, 0, -20)

        ## Clicking couple of times because sometimes whatsapp web responds after two clicks
        try:
            action.click()
            action.perform()
            action.click()
            action.perform()
        except Exception as e:
            pass
        try:
            name = browser.find_element_by_class_name("_2zCDG").text
            message = browser.find_elements_by_class_name("vW7d1")[-1]
            if name not in bot_users:
                text_box = browser.find_element_by_class_name("_2S1VP")
                response = "Hi "+name+". Extrimis here :-) . I am getting ready for you, Wait for my response, this might take a while.\n"
                text_box.send_keys(response)
                bot_users[name] = ChatBot('Extrimis')
                bot_users[name].set_trainer(ChatterBotCorpusTrainer)
                # Train the chatbot based on the english corpus
                bot_users[name].train("chatterbot.corpus.english")
            if name in bot_users and "bye" in message.text.lower():
                text_box = browser.find_element_by_class_name("_2S1VP")
                response = "Bye "+name+". It was pleasure chating with you.\n"
                text_box.send_keys(response)
                del bot_users[name]
            elif name in bot_users:
                getReply(bot_users[name], message.text)
            group_title = wait.until(EC.presence_of_element_located(( 
                By.XPATH, x_arg))) 
            group_title.click() 
        except Exception as e:
            print(e)
            pass
```
